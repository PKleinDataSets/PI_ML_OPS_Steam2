```python
import pandas as pd
```

```python
ruta_data = '/content/drive/MyDrive/Projecto Individual MLOPS/PI_ML_OPS-FT/PI MLOps - STEAM/'
```

```python
# Lee el archivo reviews Parquet en un DataFrame

ruta_archivo_reviews = ruta_data + 'reviews.parquet'
df_reviews = pd.read_parquet(ruta_archivo_reviews)
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Inicializar el analizador de sentimientos de VADER
sia = SentimentIntensityAnalyzer()

# Función para asignar valores según la escala
def get_sentiment_score(text):
    if text == '':
      return 1 # returna neutral si esta vacia
    sentiment = sia.polarity_scores(text)
    compound_score = sentiment['compound']
    if compound_score >= 0.05:
        return 2  # Positivo
    elif compound_score <= -0.05:
        return 0  # Malo
    else:
        return 1  # Neutral

# Aplicar la función de análisis de sentimiento a la columna 'review' y crear 'sentiment_analysis'
df_reviews['sentiment_analysis'] = df_reviews['review'].apply(get_sentiment_score)

# Reemplazar la columna 'review' por la nueva 'sentiment_analysis'
df_reviews.drop(columns=['review'], inplace=True)

# Mostrar las primeras filas del DataFrame actualizado
print(df_reviews.head())
```

```python
df_reviews.to_parquet('reviews_sa.parquet')
df_reviews.to_csv('reviews_sa.csv')
```
